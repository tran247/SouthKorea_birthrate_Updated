```python
import pandas as pd
import os
```

```python
#name of csv file
file = "../resources/college_edu.csv"
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file)
```

```python
df.head(100)
#Indicator = college edu
```

```python
df.count()
```

```python
#rename column
df = df.rename(columns={'LOCATION': "Country", "INDICATOR": "Indicator_edu", "SUBJECT": "Subject_edu",
                        "MEASURE": "Measure_edu", "Time_edu":"Time"})

df.head()
```

```python
#drop flag and frequency column
```

```python
#get only Korea
kor_edu1 = df.loc[df['Country'] == 'KOR']
kor_edu2 = kor_edu1[(kor_edu1['Time'] >= 1995) & (kor_edu1['Time'] <= 2018)]

kor_edu = kor_edu2.drop(['FREQUENCY', 'Flag Codes'], axis = 1)
kor_edu
```

```python
kor_edu.to_csv("kor_edu_sy.csv")
```
