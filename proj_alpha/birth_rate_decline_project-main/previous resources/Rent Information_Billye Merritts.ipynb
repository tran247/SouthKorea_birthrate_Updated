```python
#imports
import pandas as pd
import os
import numpy
```

```python
#Load Data Files
household_data = "resources/DBZ.Household_Data.csv"
```

```python
#Store DataFrame
household_data1 = pd.read_csv(household_data)
```

```python
#Print Data
household_data1.head()
```

```python
kor_inc = household_data1.loc[household_data1['LOCATION'] == 'KOR']
kor_inc.head()
```

```python
kor_debt = kor_inc[(kor_inc['TIME'] >= 1995) & (kor_inc['TIME'] <= 2018)]
kor_debt = kor_debt.drop(['FREQUENCY', 'Flag Codes'], axis = 1)
```
