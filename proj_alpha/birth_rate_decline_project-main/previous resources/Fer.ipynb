```python
import pandas as pd
```

```python
fer = pd.read_csv("Resources/fertility.csv")
```

```python
fer_df = pd.DataFrame(fer)
fer_df
```

```python
fer_kor_df = fer_df.loc[fer_df["LOCATION"]=="KOR"]
```

```python
fer_kor_df=fer_kor_df.drop(['FREQUENCY', 'Flag Codes'], axis = 1)
```

```python
fer_kor_df = fer_kor_df.rename(columns={"LOCATION": "Country", "INDICATOR": "Indicator_Fer", "SUBJECT": "Subject", "MEASURE": "Measure", "TIME":"Time"})
```

```python
fer_kor_df 
```

```python
fer_kor_df = fer_kor_df[(fer_kor_df['Time'] >= 1990) & (fer_kor_df['Time'] <= 2018)]
fer_kor_df
```

```python
fer_kor_df.to_csv(r'resources/fer_cleaned.csv',index=False)
```
