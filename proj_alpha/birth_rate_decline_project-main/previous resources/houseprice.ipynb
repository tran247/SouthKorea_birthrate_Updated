```python
import pandas as pd
import os
import csv
```

```python
houseprice_df=pd.read_csv("resources/nominalhousingprices.csv")
houseprice_df.head()
```

```python
kor_houseprice=houseprice_df.loc[houseprice_df["LOCATION"]=="KOR"]
kor_houseprice
```

```python
kor_houseprice = kor_houseprice.loc[kor_houseprice['LOCATION'] == 'KOR']
kor_houseprice = kor_houseprice[(kor_houseprice['TIME'] >= 1995) & (kor_houseprice['TIME'] <= 2018)]
kor_houseprice = kor_houseprice.drop(['FREQUENCY', 'Flag Codes'], axis = 1)
kor_houseprice
```

```python
kor_houseprice.to_csv("resources/kor_houseprice_clean.csv")
```
