```python
import pandas as pd
```

```python
#Tristan
fer_df = pd.DataFrame(pd.read_csv("resources/fer_cleaned.csv"))
fer_df = fer_df[(fer_df['Time'] >= 1995) & (fer_df['Time'] <= 2018)]
fer_df = fer_df.rename(columns={"Indicator_Fer": "Indicator"})
fer_df.head()
```

```python
#Catherine
income_df = pd.DataFrame(pd.read_csv("resources/income_cleaned.csv"))
income_df = income_df.rename(columns={"LOCATION": "Country", "INDICATOR": "Indicator", "SUBJECT": "Subject", "MEASURE": "Measure", "TIME":"Time"})
income_df
```

```python
#Akadia
houseprice_df = pd.DataFrame(pd.read_csv("resources/kor_houseprice_clean.csv"))
houseprice_df = houseprice_df.drop(['Unnamed: 0'], axis = 1)
houseprice_df = houseprice_df.rename(columns={"LOCATION": "Country", "INDICATOR": "Indicator", "SUBJECT": "Subject", "MEASURE": "Measure", "TIME":"Time"})
houseprice_df
```

```python
#Sy
edu_df = pd.DataFrame(pd.read_csv("sy/kor_edu_sy.csv"))
edu_df = edu_df.drop(['Unnamed: 0'], axis = 1)
edu_df = edu_df.rename(columns={"Indicator_edu": "Indicator", "Subject_edu": "Subject", "Measure_edu": "Measure"})
edu_df
```

```python
#Billye
debt_df = pd.DataFrame(pd.read_csv("resources/kor_debt.csv"))
debt_df = debt_df.drop(['Unnamed: 0'], axis = 1)
debt_df = debt_df.rename(columns={"LOCATION": "Country", "INDICATOR": "Indicator", "SUBJECT": "Subject", "MEASURE": "Measure", "TIME":"Time"})
debt_df
```

```python
#Concat Time, YAYYYYYY!!!!
frames = [fer_df, income_df, houseprice_df, edu_df, debt_df]
```

```python
result = pd.concat(frames)
```

```python
#result = result.reset_index()
result = result.set_index("Country")
```

```python
result
```

```python
result.to_csv("resources/database.csv")
```
