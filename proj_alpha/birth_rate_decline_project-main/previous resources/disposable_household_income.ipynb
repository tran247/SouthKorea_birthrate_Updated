```python
# Import dependencies
import pandas as pd
import numpy as np
import csv
```

```python
# Load data file 
household_income = "resources/income.csv"
```

```python
# Store as dataframe
household_income = pd.read_csv(household_income)
```

```python
# Print dataframe
household_income.head()
# HHDI == HOUSEHOLD DISPOSABLE INCOME
```

```python
kor_inc = household_income.loc[household_income['LOCATION'] == 'KOR']
kor_inc_time = kor_inc[(kor_inc['TIME'] >= 1995) & (kor_inc['TIME'] <= 2018)]
# kor_inc_time.set_index('LOCATION', inplace = True)
```

```python
kor_inc_time = kor_inc_time.drop(['FREQUENCY', 'Flag Codes'], axis = 1)
kor_inc_time
```

```python
kor_inc_time.to_csv(r'resources/income_cleaned.csv',index=False)
```
